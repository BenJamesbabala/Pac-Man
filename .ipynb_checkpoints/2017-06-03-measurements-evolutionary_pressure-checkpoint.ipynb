```python
def clear_sco(sss):
    clear_scores = []
    for file in sss:
        kk = []
        for s in file:
            kk.append(s[0])
        clear_scores.append(kk)
        
    return clear_scores
```

```python
final_scores_best = np.array(clear_sco(scores_best))
final_scores_avg = np.array(clear_sco(scores_avg))
```

```python
sco_medium = [final_scores_avg[0] / final_scores_best[0]]
sco_high = [final_scores_avg[1] / final_scores_best[1]]
```

```python
df_medium = pd.DataFrame(sco_medium, columns=list(range(len(sco_medium[0]))))
df_medium = df_medium.transpose()
df_high = pd.DataFrame(sco_high, columns=list(range(len(sco_high[0]))))
df_high = df_high.transpose()
```

```python
plt.style.use('seaborn-paper')
ax = df_medium.plot(style=['-', '-', '--'])
ax.set_ylabel('fitness', fontsize=12)
ax.set_xlabel('generations', fontsize=12)
ax.tick_params(labelsize=12)
ax.legend().set_visible(False)
plt.title('Evolutionary pressure', fontsize=14);
```
