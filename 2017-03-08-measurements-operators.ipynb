```python
import pandas as pd
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df1 = pd.read_csv('data/2017-03-08-measurements-operators/Pruebas1_clean.csv')
df2 = pd.read_csv('data/2017-03-08-measurements-operators/Pruebas2_clean.csv')
df_all = pd.concat([df1, df2], ignore_index=True)
```

```python
df_all.head()
```

Sabemos que las probabilidades de cruce y mutación, el tamaño de la poblaciónn número de iteraciones e iteraciones por individuo son siempre fijos. Además el fitness es linealmente dependiente de los puntos promedios, así que vamos a quitar estas variables porque no nos aportan nada.

```python
df = df_all[['Selección', 'Cruce', 'Mutación', 'Puntos_promedio', 'Tiempo_de_Ejecucion']]
```

```python
df.head()
```

```python
sns.lmplot(x='Tiempo_de_Ejecucion', y='Puntos_promedio', data=df, hue='Selección', fit_reg=False);
```

```python
g = sns.stripplot(x="Selección", y="Puntos_promedio", data=df, jitter=True);
g.set_xticklabels(g.get_xticklabels(), rotation=10);
```

```python
g = sns.PairGrid(df,
                 x_vars=["Selección", "Cruce", "Mutación"],
                 y_vars=["Puntos_promedio"],
                 aspect=.75, size=3.5)
g.map(sns.stripplot, palette="deep", jitter=True);
for ax in g.axes.flat:
    for label in ax.get_xticklabels():
        label.set_rotation(20)
```

```python
g = sns.PairGrid(df,
                 x_vars=["Selección", "Cruce", "Mutación"],
                 y_vars=["Puntos_promedio"],
                 aspect=.85, size=3.5)
g.map(sns.violinplot, palette="deep", inner=None);
g.map(sns.swarmplot, color="w", alpha=.5);
for ax in g.axes.flat:
    for label in ax.get_xticklabels():
        label.set_rotation(20)
```

Como curiosidad, vamos a ver el tiempo de ejecución.

```python
g = sns.PairGrid(df,
                 x_vars=["Selección", "Cruce", "Mutación"],
                 y_vars=["Tiempo_de_Ejecucion"],
                 aspect=.85, size=3.5)
g.map(sns.violinplot, palette="deep", jitter=True);
for ax in g.axes.flat:
    for label in ax.get_xticklabels():
        label.set_rotation(20)
```
