```python
df = pd.read_csv('data/2016-12-20-measurements.csv')
```

## Puntuación sobre el tiempo
La cantidad de puntos que consigue, dependiendo del tiempo que tarda en ejecutar.

```python
sns.jointplot(df['Averagepoints'], df['time'], kind='reg', xlim=(0, 15000), ylim=(0, 1500));
```

## Fitness sobre el tiempo
**Minimización** del fitness.  
La función que usamos es de la forma $f(x) = 10000 - score$.

```python
sns.jointplot(df['Fitness'], df['time'], kind='reg', xlim=(85000, 100000), ylim=(0, 1500));
```
