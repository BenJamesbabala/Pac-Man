```python
import pandas as pd
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df50 = pd.read_csv('data/2017-03-21-indices-50codons.csv')
df100 = pd.read_csv('data/2017-03-21-indices-100codons.csv')
```

```python
df50 = df50[[0, 1]]
df100 = df100[[0,1]]
```

```python
df50.head()
```

## Wrappings según el número de índices

```python
sns.set(context='talk')
sns.set_style("whitegrid")
```

### Con 50 índices

```python
plot50 = sns.pointplot(x='idx', y='num_wrappings', data=df50, markers='.', capsize=.2)
for ind, label in enumerate(plot50.get_xticklabels()):
    if ind % 2 == 0:  # every 10th label is kept
        label.set_visible(True)
    else:
        label.set_visible(False)
```

### Con 100 índices

```python
plot100 = sns.pointplot(x='idx', y='num_wrappings', data=df100, markers='', capsize=.2)
for ind, label in enumerate(plot100.get_xticklabels()):
    if ind % 5 == 0:  # every 10th label is kept
        label.set_visible(True)
    else:
        label.set_visible(False)
```
